```python
import pandas as pd
from citipy import citipy
from datetime import datetime
```

```python
file = "hurricane_data.csv"
```

```python
total_weather_data = pd.read_csv(file, dtype=str, skipinitialspace=True)
```

```python
total_weather_data.head()
```

```python
condensed_weather_data = total_weather_data.drop(columns = ["34_NE","34_SE", "34_SW", "34_NW", "50_NE","50_SE", "50_SW", "50_NW", "64_NE","64_SE", "64_SW", "64_NW", "Null" ])
```

```python
condensed_weather_data["Name"] = ""
condensed_weather_data["Storm ID"] = ""
condensed_weather_data["Start Date"] = ""
condensed_weather_data["End Date"] = ""
condensed_weather_data["Duration"] = ""
condensed_weather_data["Category"] = ""
condensed_weather_data["Nearest City"] = ""
condensed_weather_data.head()
```

```python
#strip the Direction from the Latitude and Longitude
lats = condensed_weather_data["Latitude"].str.split("([A-Z]+)", expand=True)
lons = condensed_weather_data["Longitude"].str.split("([A-Z]+)", expand=True)
#grab the number from index 0
lats = lats[0]
lons = lons[0]
#use citipy to find the nearest city
condensed_weather_data.loc[:, "Latitude"] = lats
condensed_weather_data.loc[:, "Longitude"] = lons
#change the column to numeric
condensed_weather_data["Latitude"] = pd.to_numeric(condensed_weather_data["Latitude"])
condensed_weather_data["Longitude"] = pd.to_numeric(condensed_weather_data["Longitude"])
#convert Longitude column to negative
condensed_weather_data["Longitude"] *= -1
```

```python
#use citipy to find the nearest city
latitude = condensed_weather_data["Latitude"]
longitude = condensed_weather_data["Longitude"]
coordinates = zip(latitude, longitude)
cities = []
for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    cities.append(citipy.nearest_city(lat, lon))
```

```python
#figure out the city name
city_name = []
for city in cities:
    name = city.city_name
    city_name.append(name)
condensed_weather_data.loc[:, "Nearest City"] = city_name
condensed_weather_data.head()
```
